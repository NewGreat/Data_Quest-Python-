# 6: Finding The Average Salary With The Sum And Nrow Functions

# Instructions
Calculate the average salary:
Use the sum function to find the sum of all the values in the Salary column of whiteHouse.
Use the nrow function to find the number of White House employees.
Divide the sum of the salaries by the number of employees.
Assign the result to averageSalary.


```r
numberOfEmployees <- nrow(whiteHouse["Name"])
print(numberOfEmployees)
```

```r
averageSalary <- sum(whiteHouse["Salary"]) / numberOfEmployees
print(averageSalary)
```

# 9: Finding The Indexes For A Vector's Minimum And Maximum Values

# Instructions
Find the White House employee with the highest salary and assign the result to highestEarner.
Hint
which.max(whiteHouse[,"Salary"]) returns the index you need, which you can then use to select the row you want.

```r
minimumindex <- which.min(whiteHouse[,"Salary"])
print(minimumindex)
```

```r
minimumSalaryRow <- whiteHouse[minimumindex,]
print(minimumSalaryRow)
```

```r
lowestEarner <- minimumSalaryRow["Name"]
print(lowestEarner)
```

```r
maximumIndex <- which.max(whiteHouse[,"Salary"])
print(maximumIndex)
```

```r
maximumSalaryRow <- whiteHouse[maximumIndex,]
print(maximumSalaryRow)
```

```r
highestEarner <- maximumSalaryRow["Name"]
print(highestEarner)
```
