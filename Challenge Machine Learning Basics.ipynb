# 1. Data Cleaning
Instructions Remove all rows where the value for horsepower is ? and convert the horsepower column to a float.
Assign the new Dataframe to filtered_cars.
Hint
Use the Series method astype to convert the column to float.

```python
import pandas as pd
columns = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", " car name"]
cars = pd.read_table('auto-mpg.data.txt',names=columns, delim_whitespace=True)
cars
```

```python
filtered_cars = cars[cars['horsepower'] != '?']
filtered_cars.info()
```

```python
filtered_cars['horsepower'] = filtered_cars['horsepower'].astype("float")
filtered_cars.info()
```

# 2. Data Exploration
Instructions Use the Dataframe plot to generate a grid of subplots (2 rows by 1 column):
On the top plot, generate a scatter plot with the horsepower column on the x-axis and the mpg column on the y-axis.
On the bottom plot, generate a scatter plot with the weight column on the x-axis and the mpg column on the y-xis.
Hint
Use the Dataframe method plot to generate simple scatter plots. Documentation.

```python
import matplotlib.pyplot as plt
filtered_cars.plot('horsepower', 'mpg', kind='scatter', c='red')
filtered_cars.plot('weight', 'mpg', kind='scatter', c='blue')
#plt.show()
```

# 3. Fitting A Model
Instructions Create a new instance of the LinearRegression model and assign it to lr.
Use the fit method to fit a linear regression model using the horsepower column as the input.
Use the model to make predictions on the same data the model was trained on (the horsepower column from filtered_cars) and assign the resulting predictions to predictions.
Display the first 5 values in predictions and the first 5 values in the mpg column from filtered_cars.
Hint
Don't forget to import LinearRegression from sklearn.linear_model.

```python
import sklearn
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(filtered_cars[["horsepower"]], filtered_cars["mpg"])

predictions = lr.predict(filtered_cars[["horsepower"]])
print(predictions[0:5])
print(filtered_cars["horsepower"][0:5])
```

# 4. Plotting The Predictions
Instructions Generate 2 scatter plots on the same chart (Matplotlib axes instance):
One containing the horsepower values on the x-axis against the predicted fuel efficiency values on the y-axis. Use blue for the color of the dots.
One containing the horsepower values on the x-axis against the actual fuel efficiency values on the y-axis. Use red for the color of the dots.
Hint
You can generate 2 scatter plots on one just by calling plt.scatter() twice and without creating any Axes objects manually.

```python
plt.scatter(filtered_cars["horsepower"], predictions, c='red')
plt.scatter(filtered_cars["horsepower"], filtered_cars["mpg"], c='blue')
#plt.show()
```

# 5. Error Metrics
Instructions Calculate the MSE of the predicted values and assign to mse.
Calculate the RMSE of the predicted values and assign to rmse.
Hint
Recall that you need to import the mean_squared_error function from sklearn.metrics.

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(filtered_cars["mpg"], predictions)
print(mse)
```

```python
rmse = (mse) ** (1/2)
print(rmse)
```
