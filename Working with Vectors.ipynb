# 7. Multiplying A Matrix By A Vector
Instructions
Multiply nba_rows by nba_coefs.
nba_rows has two columns. The first contains the number of field goals each player attempted in 2013, and the second is a constant 1 value that enables us to add in the intercept.
Assign the result to predictions.
Hint
Remember to use np.dot.

```python
import numpy as np

coefs = np.array([[3], [-1]])
rows = np.array([[2,1], [5,1], [-1,1]])

np.dot(rows, coefs)

predictions = np.dot(rows, coefs)
print(predictions)
```

```python
nba_coefs = np.array([[slope], [intercept]])
nba_rows = np.vstack([nba["Weight"], np.ones(nba.shape[0])]).T

predictions = np.dot(nba_rows, nba_coefs)
print(predictions)
```
